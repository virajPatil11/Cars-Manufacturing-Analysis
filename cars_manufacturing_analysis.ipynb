```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("usa_cars_datasets.csv")
```

```python
df.head()
```

```python
df['brand'].unique()
```

```python
for country in df['brand'].unique():
  print(country, len(df[df['brand']== country]))
```

## 1.TOP 5 CAR BRANDS IN TERM OF NUMBER OF CARS

```python
data = []

for country in df['brand'].unique():
    data.append([country, len(df[df['brand'] == country])])
    
data = pd.DataFrame(data,columns = ['brand', 'freq'])
data = data.sort_values(by = 'freq',ascending = False).head()
```

```python
data
```

```python
data.plot(x  = 'brand', y = 'freq', kind = 'bar', figsize = (10,5))
```

## 2. GRAPH ON CAR BRANDS BASED ON FREQUENCY OF CARS

```python
data = []

for country in df['brand'].unique():
    data.append([country, len(df[df['brand'] == country])])
    
data = pd.DataFrame(data,columns = ['brand', 'freq'])
data = data.sort_values(by = 'freq',ascending = False)
data.plot(x  = 'brand', y = 'freq', kind = 'bar', figsize = (10,5))
```

### 3.CAR INDUSTRY TREND

```python
data = []

for country in df['year'].unique():
    data.append([country, len(df[df['year'] == country])])
    
data = pd.DataFrame(data,columns = ['year', 'freq'])
data = data.sort_values(by = 'freq',ascending = True)
data.plot(x  = 'year', y = 'freq', kind = 'line', figsize = (10,5))
```

## 4. Most Popular Car Colours

```python
data = []

for country in df['color'].unique():
    data.append([country, len(df[df['color'] == country])])
    
data = pd.DataFrame(data,columns = ['color', 'freq'])
data = data.sort_values(by = 'freq',ascending = False)
data.plot(x  = 'color', y = 'freq', kind = 'bar', figsize = (10,5))
```

## 5. Which car brand is having maximum average price

```python
df.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar(figsize = (10,5))
```

## 6.Which State is having most expensive cars

```python
plt.figure(figsize = (20,10))
df.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar()
plt.show()
```
